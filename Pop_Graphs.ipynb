```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = 'c:\\Users\\Tang\\OneDrive - University of Bristol\\MSc project\\Project\\Causal_Relationship_Inferring\\'
```

```python
filename = 'output_pops_A.csv'
st = 1000
file = path+filename
pops = np.genfromtxt(filename, delimiter=',')
pops = pops[st:,:]
Nsp = np.shape(pops)[1]
colours = ['g', 'r', 'b', 'y', 'c', 'k', 'm', 'o']

for s in range(Nsp):
    plt.plot(pops[:,s], c=colours[s])

plt.show()
```
